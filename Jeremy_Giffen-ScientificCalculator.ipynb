```python
import math
class SciCalculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
        
    def addition(num1, num2):
        print("Addition", num1+num2)
        
    def subtraction(num1, num2):
        print("Subtraction", num1-num2)
        
    def multiplication(num1, num2):
        print("Multiplication: ", num1*num2)
        
    def division(num1, num2):
        print("Division: ", num1/num2)
    
    def sqrt(num1):
        print("Square root: ", math.sqrt(num1))
        
    def cbrt(num1):
        print("Cube root: ", math.cbrt(num1))
        
    def floordivision(num1, num2):
        print("Floor Division: ", num1//num2)
        
    def modulus(num1, num2):
        print("Modulus: ", num1%num2)
        
    def exponent(num1, num2):
        print("Exponentation: ", num1**num2)
        
    def absvalue(num1):
        print("Absolute value: ", abs(num1))
        
    def trigsin(num1):
        print("Sine: ", math.sin(math.radians(num1)))
        
    def trigcos(num1):
        print("Cosine: ", math.cos(math.radians(num1)))
    
    def trigtan(num1):
        print("Tangent: ", math.tan(math.radians(num1)))
        
    def logarithm(num1):
        print("Logarithm: ", math.log(num1))
        
    def gcd(num1, num2):
        print("Greatest common divisor: ", math.gcd(num1, num2))
    
    def lcm(num1, num2):
        print("Least common multiple: ", num1*num2 // math.gcd(num1, num2))
        
    def hypersin(num1):
        print("Hyperbolic Sine: ", math.sinh(num1))
        
    def hypercos(num1):
        print("Hyperbolic Cosine: ", math.cosh(num1))
    
    def hypertan(num1):
        print("Hyperbolic Tangent: ", math.tanh(num1))
    
    
while True:
    print("1: Addition")
    print("2: Subtraction")
    print("3: Multiplication")
    print("4: Division")
    print("5: Square root")
    print("6: Cube root")
    print("7: Floor division")
    print("8: Modulus")
    print("9: Exponent")
    print("10: Absolute value")
    print("11: Trigonometry Sine")
    print("12: Trigonometry Cosine")
    print("13: Trigonometry Tangent")
    print("14: Logarithm")
    print("15: Greatest common divisor")
    print("16: Least common multiple")
    print("17: Hyperbolic Sine")
    print("18: Hyperbolic Cosine")
    print("19: Hyperbolic Tangent")
    print("20: Exit")
    choice= int(input("Enter your choice (1-20):"))
    
    if choice==1:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.addition(num1, num2)
    elif choice==2:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.subtraction(num1, num2)
    elif choice==3:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.multiplication(num1, num2)
    elif choice==4:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        
        if num2==0:
            print("Infinity")
        
        else:
            SciCalculator.division(num1, num2)
    elif choice==5:
        num1=int(input("Enter a number"))
        SciCalculator.sqrt(num1)
    elif choice==6:
        num1=int(input("Enter a number"))
        SciCalculator.cbrt(num1)
    elif choice==7:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.floordivision(num1, num2)
    elif choice==8:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.modulus(num1, num2)   
    elif choice==9:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.exponent(num1, num2)   
    elif choice==10:
        num1=int(input("Enter a number: "))
        SciCalculator.absvalue(num1)
    elif choice==11:
        num1=int(input("Enter a number: "))
        SciCalculator.trigsin(num1)
    elif choice==12:
        num1=int(input("Enter a number: "))
        SciCalculator.trigcos(num1)
    elif choice==13:
        num1=int(input("Enter a number: "))
        SciCalculator.trigtan(num1)
    elif choice==14:
        num1=int(input("Enter a number: "))
        SciCalculator.logarithm(num1)
    elif choice==15:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.gcd(num1, num2)
    elif choice==16:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        SciCalculator.lcm(num1, num2)
    elif choice==17:
        num1=int(input("Enter a number: "))
        SciCalculator.hypersin(num1)
    elif choice==18:
        num1=int(input("Enter a number: "))
        SciCalculator.hypercos(num1)
    elif choice==19:
        num1=int(input("Enter a number: "))
        SciCalculator.hypertan(num1)
    elif choice==20:
        break
    else:
        print("Invalid choice.  Choose a number between 1 - 20")
```
